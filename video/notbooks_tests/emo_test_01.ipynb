```python
import cv2
from deepface import DeepFace

# Open the webcam
cap = cv2.VideoCapture(0)
try:
    while True:
        # Capture frame-by-frame
        ret, frame = cap.read()
        
        # Analyzing facial expression
        result = DeepFace.analyze(frame, actions=['emotion'])

            # Access emotion from the list (if result is a list)
        if isinstance(result, list):
            emotion = result[0]["dominant_emotion"]  # Assuming emotion is in the first element
        else:
            emotion = result["dominant_emotion"]  # If it's a dictionary

        # Printing the result
        print(f"The dominant emotion is: {emotion}")
        
        #debug print(result)
    
        # Setup status box
        cv2.rectangle(frame, (0,0), (225,73), (245,117,16), -1)
        
        # Stage data
        cv2.putText(frame, 'Emotion:', (65,12), 
        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        cv2.putText(frame, emotion, 
        (60,60), 
        cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 0, 0), 2, cv2.LINE_AA)
        
        # Display the resulting frame
        
        cv2.imshow('Webcam Feed', frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
except Exception as e:
    print(e)

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
